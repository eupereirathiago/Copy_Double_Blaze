```python
import os
import time
import random
import google.generativeai as genai

# --- FUNÇÕES AUXILIARES ---
def geradorDeFrase():

    try:
      #atribuindo a chave a uma variavel
      GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
      genai.configure(api_key=GOOGLE_API_KEY)

      #selecionando qual modelo usar
      gemini_model = genai.GenerativeModel('gemini-2.5-flash')

      #atribuindo a frase a uma variavel
      response = gemini_model.generate_content("Gere somente uma frase curta dando esporro em alguem que perdeu em um jogo de azar porem ele deve continuar")

      return (response.text)

    except Exception as e:
        # caso falhe, mostra uma frase padrão
        return "Mais sorte na próxima vez, campeão! Tenta mais uma vez..."

frase = geradorDeFrase()
# print(frase)

def realizar_deposito():

    resposta = input('\nVamos fazer um deposito? (S/N) ')

    if (resposta == 's') or (resposta == 'S'):

        try:
            recarga = float(input('\nValor do depósito: R$ '))
            if recarga > 0:
                saldo = recarga
                print(f'\nSeu saldo é R$ {saldo:.2f}')
                return saldo
            else:
                print("O valor do depósito deve ser maior que zero.")
                return 0

        except ValueError:
            print("Erro: Por favor, insira um valor numérico válido.")
            return 0


    elif (resposta == 'n') or (resposta == 'N'):
        print('\nQue pena. Até a próxima!')
        return 0
    else:
        print("Opção inválida. Tente novamente.")
        return 0

def pegar_aposta(saldo_disponivel):

    try:
        aposta = float(input('\nValor da aposta: R$ '))
        if (aposta > 0) and (aposta <= saldo_disponivel):
            return aposta
        else:
            print(f"Aposta inválida. Você deve apostar um valor entre R$ 0.01 e R$ {saldo_disponivel:.2f}.")
            return 0

    except ValueError:
        print("Erro: Por favor, insira um valor numérico válido.")
        return 0

def escolher_cor():

    print('\nEscolha a sua cor:')
    print('1 - PRETO (ganho de 2x)')
    print('2 - VERMELHO (ganho de 2x)')
    print('3 - BRANCO (ganho de 14x)')

    try:
        opcao_numero = int(input('\nDigite o número da cor (1, 2 ou 3): '))
        if opcao_numero == 1:
            return 'PRETO'
        elif opcao_numero == 2:
            return 'VERMELHO'
        elif opcao_numero == 3:
            return 'BRANCO'
        else:
            print("Opção inexistente! Tente novamente.")
            return ""

    except ValueError:
        print("Erro: Isso não é um número válido! Tente novamente.")
        return ""

# --- FUNÇÃO PRINCIPAL DO JOGO ---

def jogo():

    caixa_da_casa = 0.0

    # 1. DEPÓSITO INICIAL
    saldo = realizar_deposito()
    if saldo <= 0:
        return

    # Loop principal do jogo (uma rodada por vez)
    while saldo > 0:
        resposta_aposta = input('\nVamos apostar? (S/N) ')

        if (resposta_aposta == 's') or (resposta_aposta == 'S'):
            valor_aposta = pegar_aposta(saldo)
            if valor_aposta == 0:
                continue
            opcaoEscolhida_cor = escolher_cor() # Pegar uma escolha de cor válida

            if not opcaoEscolhida_cor: # Para verificar se a string não está vazia
                continue

            print(f'\nVocê apostou R$ {valor_aposta:.2f} na cor {opcaoEscolhida_cor}.')
            print('Sorteando...')
            time.sleep(5)

            # Sorteio
            opcoes = ['PRETO', 'VERMELHO', 'BRANCO']

            # O truque para que o playernao ganhe
            # Se a casa nao tiver 10,00 e ele acerter e sorteado outra cor hehe...
            # É criado uma lista somente com as cores nnao escolhidas
            if caixa_da_casa < 10.00:
                opcoes_de_derrota = [cor for cor in opcoes if cor != opcaoEscolhida_cor]
                resultado_sorteio = random.choice(opcoes_de_derrota)
            else:
                resultado_sorteio = random.choices(opcoes, weights=[7, 7, 1], k=1)[0]

            print(f'A cor sorteada foi: {resultado_sorteio}!')


            # Lógica de resultado e pagamento
            if opcaoEscolhida_cor == resultado_sorteio:

                # regra do branco 14x
                ganho = valor_aposta if resultado_sorteio != 'BRANCO' else valor_aposta * 13
                saldo += ganho
                caixa_da_casa -= ganho
                print(f"PARABÉNS! Você ganhou R$ {ganho:.2f}!")
            else:
                saldo -= valor_aposta
                caixa_da_casa += valor_aposta
                print(f"Que pena! Você perdeu R$ {valor_aposta:.2f}.")
                print(f"{geradorDeFrase()}\n")

            print(f'Seu novo saldo é: R$ {saldo:.2f}')

        elif (resposta_aposta == 'n') or (resposta_aposta == 'N'):
            print(f"\nOk! Você está saindo com R$ {saldo:.2f}. Até a próxima!")
            break

        else:
            print("Opção inválida. Digite 'S' ou 'N'.")
            continue

        if saldo <= 0:
            print("\nVocê ficou sem saldo. Fim de jogo!")
            break

# --- PONTO DE ENTRADA DO PROGRAMA ---

def main():
    print(' ########## BEM VINDO AO SORTE NO AZAR ########## ')
    print('--------------------------------------------------')
    jogo()

if __name__ == "__main__":
    main()
```
